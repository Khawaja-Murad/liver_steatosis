```python
# === 0) Imports & Config ===
import os, copy, time
import numpy as np, pandas as pd
import scipy.io
from PIL import Image
import torch, torch.nn as nn, torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, WeightedRandomSampler
import torchvision.transforms as T
from torchvision.models import resnet34, ResNet34_Weights
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, mean_absolute_error, r2_score, classification_report
)
import matplotlib.pyplot as plt

# reproducibility
seed = 42
np.random.seed(seed)
torch.manual_seed(seed)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"[INFO] Device: {device}")

# === 1) Load Data & Hold‑out Test Split ===
mat = scipy.io.loadmat('/kaggle/input/dataset-of-bmode-fatty-liver-ultrasound-images/dataset_liver_bmodes_steatosis_assessment_IJCARS.mat')
entries = mat['data'][0]

vols, labels, fats = [], [], []
for e in entries:
    vols.append(e['images'])                    # (S,H,W)
    labels.append(int(e['class'].squeeze()))
    fats.append(float(e['fat'].squeeze()))
vols   = np.stack(vols)                         # (N,S,H,W)
labels = np.array(labels, dtype=np.int64)
fats   = np.array(fats, dtype=np.float32)

# split out final test set (20%), stratified on label
idx = np.arange(len(labels))
trainval_idx, test_idx = train_test_split(
    idx, test_size=0.2, stratify=labels, random_state=seed
)
vols_tv, vols_test   = vols[trainval_idx], vols[test_idx]
lab_tv,  lab_test    = labels[trainval_idx], labels[test_idx]
fat_tv,  fat_test    = fats[trainval_idx],   fats[test_idx]

print(f"[INFO] Total: {len(labels)}, TrainVal: {len(vols_tv)}, Test: {len(vols_test)}")

# === 2) Dataset & Slice Picker ===
class LiverDataset(Dataset):
    def __init__(self, vols, labs, fats, scaler, transform, train=True):
        self.vols  = vols
        self.labs  = labs
        self.fats  = fats
        self.scaler = scaler
        self.tf    = transform
        self.train = train
        if train:
            self.fats_s = scaler.fit_transform(fats.reshape(-1,1)).flatten()
        else:
            self.fats_s = scaler.transform(fats.reshape(-1,1)).flatten()
    def __len__(self): return len(self.vols)
    def __getitem__(self, i):
        vol = self.vols[i]                      # (S,H,W)
        sl = np.random.randint(vol.shape[0]) if self.train else vol.shape[0]//2
        img2d = vol[sl].astype(np.float32)
        img2d = (img2d - img2d.min())/ (img2d.ptp()+1e-8) *255
        img2d = img2d.astype(np.uint8)
        img = Image.fromarray(img2d, mode='L')
        img = self.tf(img)
        return img, self.labs[i], self.fats_s[i], self.fats[i]

# === 3) Losses & Model ===
class FocalLoss(nn.Module):
    def __init__(self, gamma=2.0, alpha=None): 
        super().__init__()
        self.gamma=gamma
        self.alpha=alpha
    def forward(self, logits, targets):
        ce = F.cross_entropy(logits, targets, weight=self.alpha, reduction='none')
        pt = torch.exp(-ce)
        return ((1-pt)**self.gamma * ce).mean()

class HybridNet(nn.Module):
    def __init__(self, num_classes=2):
        super().__init__()
        self.back = resnet34(weights=ResNet34_Weights.IMAGENET1K_V1)
        in_f = self.back.fc.in_features
        self.back.fc = nn.Identity()
        self.fc_cls = nn.Sequential(
            nn.Linear(in_f, 256), nn.ReLU(inplace=True), nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )
        self.fc_reg = nn.Sequential(
            nn.Linear(in_f, 512), nn.SiLU(inplace=True), nn.Dropout(0.3),
            nn.Linear(512, 256), nn.SiLU(inplace=True), nn.Dropout(0.3),
            nn.Linear(256,1)
        )
    def forward(self,x):
        f = self.back(x)
        return self.fc_cls(f), self.fc_reg(f).squeeze(1)

# === 4) Plot Helpers ===
def plot_cm(cm, name):
    plt.figure(figsize=(4,4)); plt.imshow(cm, cmap='Blues')
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j,i,cm[i,j], ha='center',va='center',
                     color='white' if cm[i,j]>cm.max()/2 else 'black')
    plt.title(name); plt.xlabel('Pred'); plt.ylabel('True')
    plt.tight_layout(); plt.savefig(f'{name.replace(" ","_")}.png'); plt.show(); plt.close()

def plot_reg(y,t,name):
    plt.figure(figsize=(5,5)); plt.scatter(t,y,alpha=0.6)
    m,n = min(y.min(),t.min()), max(y.max(),t.max())
    plt.plot([m,n],[m,n],'r--'); plt.title(name)
    plt.xlabel('True Fat%'); plt.ylabel('Pred Fat%')
    plt.tight_layout(); plt.savefig(f'{name.replace(" ","_")}.png'); plt.show(); plt.close()

# === 5) EarlyStopping ===
class EarlyStopping:
    def __init__(self, patience=15, min_delta=1e-4):
        self.patience, self.min_delta = patience, min_delta
        self.best, self.counter = np.inf, 0
        self.should_stop=False
    def __call__(self, val_loss):
        if val_loss < self.best - self.min_delta:
            self.best, self.counter = val_loss, 0
        else:
            self.counter+=1
            if self.counter>=self.patience: self.should_stop=True

# === 6) Stratified K‑Fold on TrainVal ===
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)
best_f1, best_state, best_fold = -1, None, -1

for fold, (tr, vl) in enumerate(skf.split(vols_tv, lab_tv),1):
    print(f"\n>>> Fold {fold}")
    Vtr, Vvl = vols_tv[tr], vols_tv[vl]
    Ltr, Lvl = lab_tv[tr], lab_tv[vl]
    Ftr, Fvl = fat_tv[tr], fat_tv[vl]

    cnt = np.bincount(Ltr)
    sw = (1.0/cnt)[Ltr]
    sampler = WeightedRandomSampler(sw, len(sw), True)
    alpha = torch.tensor((len(Ltr)/(2*cnt)).astype(np.float32), device=device)

    tf_tr = T.Compose([T.RandomHorizontalFlip(), T.RandomRotation(15),
                       T.ToTensor(), T.Lambda(lambda t:t.repeat(3,1,1)),
                       T.Normalize([0.5]*3,[0.5]*3)])
    tf_vl = T.Compose([T.ToTensor(), T.Lambda(lambda t:t.repeat(3,1,1)),
                       T.Normalize([0.5]*3,[0.5]*3)])

    scaler = MinMaxScaler()
    ds_tr = LiverDataset(Vtr,Ltr,Ftr,scaler,tf_tr,train=True)
    ds_vl = LiverDataset(Vvl,Lvl,Fvl,scaler,tf_vl,train=False)

    loader_tr = DataLoader(ds_tr, batch_size=8, sampler=sampler, num_workers=2)
    loader_vl = DataLoader(ds_vl, batch_size=8, shuffle=False, num_workers=2)

    model = HybridNet().to(device)
    pri = cnt/cnt.sum()
    model.fc_cls[-1].bias.data = torch.log(torch.tensor(pri+1e-8, dtype=torch.float32)).to(device)
    crit_cls = FocalLoss(gamma=2.0, alpha=alpha)
    crit_reg = nn.MSELoss()
    opt      = torch.optim.Adam(model.parameters(), lr=1e-4)
    sched    = torch.optim.lr_scheduler.ReduceLROnPlateau(opt, mode='min', patience=5, verbose=True)
    early    = EarlyStopping(patience=15)

    history = []
    for epoch in range(1,101):
        model.train()
        tloss,tacc,tmae=0,0,0
        for imgs,lbs,fs_s,fs in loader_tr:
            imgs,lbs,fs_s=imgs.to(device),lbs.to(device),fs_s.to(device)
            opt.zero_grad()
            logits,preds=model(imgs)
            loss=crit_cls(logits,lbs)+crit_reg(preds,fs_s)
            loss.backward();opt.step()
            tloss+=loss.item()*imgs.size(0)
            tacc+=(logits.argmax(1)==lbs).sum().item()
            po=scaler.inverse_transform(preds.detach().cpu().numpy().reshape(-1,1)).flatten()
            tmae+=np.sum(np.abs(po-fs.numpy()))
        ntr=len(ds_tr)
        train_loss,train_acc,train_mae=tloss/ntr, tacc/ntr, tmae/ntr

        model.eval()
        vloss,vacc,vmae=0,0,0
        all_p,all_l,all_pr,all_tr=[],[],[],[]
        with torch.no_grad():
            for imgs,lbs,fs_s,fs in loader_vl:
                imgs,lbs,fs_s=imgs.to(device),lbs.to(device),fs_s.to(device)
                logits,preds=model(imgs)
                loss=crit_cls(logits,lbs)+crit_reg(preds,fs_s)
                vloss+=loss.item()*imgs.size(0)
                vacc+=(logits.argmax(1)==lbs).sum().item()
                all_p.extend(logits.argmax(1).cpu().numpy())
                all_l.extend(lbs.cpu().numpy())
                po=scaler.inverse_transform(preds.cpu().numpy().reshape(-1,1)).flatten()
                all_pr.extend(po);all_tr.extend(fs.numpy())
                vmae+=np.sum(np.abs(po-fs.numpy()))
        nvl=len(ds_vl)
        val_loss,val_acc,val_mae=vloss/nvl,vacc/nvl,vmae/nvl

        history.append({'epoch':epoch,'train_loss':train_loss,'train_acc':train_acc,'train_mae':train_mae,
                        'val_loss':val_loss,'val_acc':val_acc,'val_mae':val_mae,'lr':opt.param_groups[0]['lr']})
        print(f"Fold{fold} E{epoch}: TrL{train_loss:.4f} Acc{train_acc:.3f} MAE{train_mae:.3f} | VlL{val_loss:.4f} Acc{val_acc:.3f} MAE{val_mae:.3f}")

        sched.step(val_loss); early(val_loss)
        if early.should_stop: print("[INFO] Early stopping"); break

    pd.DataFrame(history).to_csv(f'history_fold{fold}.csv',index=False)
    cm=confusion_matrix(all_l,all_p)
    plot_cm(cm,f'Fold{fold} Val CM')
    plot_reg(np.array(all_pr),np.array(all_tr),f'Fold{fold} Val Reg')
    f1m=f1_score(all_l,all_p,average='macro',zero_division=0)
    pd.DataFrame([{'fold':fold,'val_acc':val_acc,'val_f1':f1m,'val_mae':val_mae}]).to_csv(f'val_metrics_fold{fold}.csv',index=False)
    if f1m>best_f1: best_f1,best_state,best_fold=f1m,copy.deepcopy(model.state_dict()),fold

# === 7) Evaluate Best Model on Test Set ===
print(f"\n[INFO] Best Fold: {best_fold} with F1={best_f1:.3f}")
torch.save(best_state,'best_model.pth')
model.load_state_dict(best_state)

tf_test=T.Compose([T.ToTensor(),T.Lambda(lambda t:t.repeat(3,1,1)),T.Normalize([0.5]*3,[0.5]*3)])
ds_test=LiverDataset(vols_test,lab_test,fat_test,scaler,tf_test,train=False)
loader_test=DataLoader(ds_test,batch_size=8,shuffle=False,num_workers=2)

tp,tl,trp,trt=[],[],[],[]
with torch.no_grad():
    for imgs,lbs,fs_s,fs in loader_test:
        imgs,lbs=imgs.to(device),lbs.to(device)
        logits,preds=model(imgs)
        tp.extend(logits.argmax(1).cpu().numpy())
        tl.extend(lbs.cpu().numpy())
        po=scaler.inverse_transform(preds.cpu().numpy().reshape(-1,1)).flatten()
        trp.extend(po);trt.extend(fs.numpy())

acc_t,f1_t,mae_t=accuracy_score(tl,tp),f1_score(tl,tp,average='macro'),mean_absolute_error(trt,trp)
cm_t=confusion_matrix(tl,tp)
print(f"[TEST] Acc {acc_t:.3f} F1 {f1_t:.3f} MAE {mae_t:.3f}")
plot_cm(cm_t,'Test CM'); plot_reg(np.array(trp),np.array(trt),'Test Reg')
pd.DataFrame([{'test_acc':acc_t,'test_f1':f1_t,'test_mae':mae_t}]).to_csv('test_metrics.csv',index=False)

```

```python
# === 8) Inference Function ===
import PIL
def inference(volume_3d, true_label, true_fat, model, scaler, transform):
    model.eval()
    sl=volume_3d.shape[0]//2
    img2d=volume_3d[sl].astype(np.float32)
    img2d=(img2d-img2d.min())/(img2d.ptp()+1e-8)*255
    img2d=img2d.astype(np.uint8)
    img=PIL.Image.fromarray(img2d,mode='L')
    img_tensor=transform(img).unsqueeze(0).to(device)
    with torch.no_grad():
        logits,pred_s=model(img_tensor)
    pred_class=logits.argmax(1).item()
    pred_f_norm=pred_s.item()
    pred_f = scaler.inverse_transform([[pred_f_norm]])[0][0]
    print("\n=== Inference ===")
    print(f"True Class: {true_label}, Pred Class: {pred_class}")
    print(f"True Fat%: {true_fat:.2f}, Pred Fat%: {pred_f:.2f}\n")
    plt.figure(figsize=(6,6)); plt.imshow(img, cmap='gray'); plt.axis('off')
    plt.title(f"T_cls:{true_label} P_cls:{pred_class} | T_fat:{true_fat:.1f}% P_fat:{pred_f:.1f}%")
    plt.show()

# Example:
inference(vols_test[1], lab_test[1], fat_test[1], model, scaler, tf_test)
```
